```python
import matplotlib as mpl        
import matplotlib.pyplot as plt 
import cartopy.crs as ccrs
import cartopy.io.img_tiles as cimgt
```

```python
# OpenStreetMap
#request = cimgt.OSM()
# Create a MapQuest open aerial instance.
#request = cimgt.MapQuestOpenAerial()
# Create a Stamen Terrain instance.
request = cimgt.Stamen('terrain-background')

extent = [-49, -48, -29, -24]
fig = plt.figure(figsize=[15,15])
ax = fig.add_subplot(1, 1, 1)  # specify (nrows, ncols, axnum)

ax = plt.axes(projection=request.crs)
ax.set_extent(extent)

ax.add_image(request, 8)
plt.show()
```

```python
from cartopy.io.shapereader import Reader
from cartopy.io.shapereader import gshhs
from cartopy.feature import ShapelyFeature
from cartopy.feature import LAND
from cartopy.feature import OCEAN
from cartopy.feature import COASTLINE
from cartopy.feature import NaturalEarthFeature
```

```python
fname = costao_file

fig = plt.figure(figsize=[15,15])
ax = fig.add_subplot(1, 1, 1)  # specify (nrows, ncols, axnum)

ax = plt.axes(projection=request.crs)
extent = [-48.7, -48.3, -27.5, -27]
ax.set_extent(extent)
# ax.add_image(request, 11)
ax.add_geometries(Reader(fname).geometries(),
                  ccrs.PlateCarree(),
                  facecolor='white', hatch='\\\\\\\\\\', alpha=0.5)
# ax.add_feature(NaturalEarthFeature('physical', 'land', '10m', edgecolor='face', facecolor='g'))
plt.show()
```
